```python
## 1. Import Libraries & Load Synthetic Titanic-like Dataset
import numpy as np
import matplotlib.pyplot as plt
import heapq
import time
import json
```

```python
# Create a random maze grid
np.random.seed(42)
rows, cols = 20, 20
maze = np.zeros((rows, cols))
```

```python
# Add random walls
for _ in range(120):
    r = np.random.randint(0, rows)
    c = np.random.randint(0, cols)
    maze[r][c] = 1  # wall

start, end = (0, 0), (19, 19)
maze[start] = 0
maze[end] = 0
```

```python
# Heuristic function (Manhattan Distance)
def heuristic(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1])
```

```python
# A* Algorithm
def a_star(maze, start, end):
    pq = []
    heapq.heappush(pq, (0, start))

    g_score = {start: 0}
    f_score = {start: heuristic(start, end)}
    parent = {}

    while pq:
        _, current = heapq.heappop(pq)

        if current == end:
            path = []
            while current in parent:
                path.append(current)
                current = parent[current]
            path.append(start)
            return path[::-1]

        r, c = current
        neighbors = [(r+1, c), (r-1, c), (r, c+1), (r, c-1)]

        for nbr in neighbors:
            nr, nc = nbr

            if 0 <= nr < rows and 0 <= nc < cols and maze[nr][nc] == 0:
                new_g = g_score[current] + 1

                if nbr not in g_score or new_g < g_score[nbr]:
                    parent[nbr] = current
                    g_score[nbr] = new_g
                    f_score[nbr] = new_g + heuristic(nbr, end)

                    heapq.heappush(pq, (f_score[nbr], nbr))

    return None
```

```python
# Run and time A*
t0 = time.time()
path = a_star(maze, start, end)
t1 = time.time()

print("â±ï¸ A* Time (seconds):", round(t1 - t0, 5))
print("âœ… Path Found:", bool(path))
```

```python
# Visualize path
if path:
    maze_copy = maze.copy()
    for r, c in path:
        maze_copy[r][c] = 2  # mark path

    plt.figure(figsize=(6,6))
    plt.imshow(maze_copy)
    plt.scatter(start[1], start[0], s=120, marker='o')
    plt.scatter(end[1], end[0], s=120, marker='X')
    plt.title("Maze Solved with A*")
    plt.axis('off')
    plt.show()
```

```python
# Log results to JSON for repo usage
log = {
    "algorithm": "A* (Manhattan)",
    "path_length": len(path) if path else None,
    "time_seconds": round(t1 - t0, 5),
    "grid_size": f"{rows}x{cols}",
    "found": bool(path)
}
```

```python
with open("Log.json", "w") as f:
    json.dump(log, f, indent=2)

print("\nðŸ“ Log saved as Log.json")
```
